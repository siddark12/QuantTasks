```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import math
timesteps = 30
starting_price = 600
volatility = 2/100
max_simulations = 5000
price_predict=[]
for i in range(max_simulations):
    
    returns=np.random.uniform(1-volatility,1+volatility,timesteps)
    price=[600]
    for ratio in returns:
        price.append(price[-1]*ratio)
    plt.plot(price)
    price_predict.append(price[-1])
    plt.ylabel('Stock Price')
    plt.xlabel('Time')
    plt.title('Paths by MonteCarlo Simultation')

plt.savefig('Paths by MonteCarlo Simultation')
plt.show()

plt.hist(price_predict, bins=100)
plt.ylabel('No. of occurrences')
plt.xlabel('Price')
plt.title('Distribution of final Stock Price')
plt.savefig('Distribution')

plt.show()
```
